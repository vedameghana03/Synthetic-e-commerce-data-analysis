# Feature Engineering

### Effective revenue

```python
df['Effective_Revenue'] = df['Revenue'] - df['Ad_Spend']
```

```python
df[['Revenue', 'Ad_Spend', 'Effective_Revenue']].head()
```

### Revenue per unit

```python
df['Revenue_per_Unit'] = df['Revenue'] / df['Units_Sold'].where(df['Units_Sold'] != 0)
```

```python
df['Revenue_per_Unit'].head()
```

### Discount per unit

```python
df['Discount_Per_Unit'] = df['Discount_Applied'] / df['Units_Sold'].where(df['Units_Sold'] != 0)
```

### CTR_Band

```python
df['Discount_Per_Unit'].head()
```

```python
def ctr_band(c):
    if c < 0.3:
        return 'Low'
    elif c <= 0.6:
        return 'Medium'
    else:
        return 'High'

df['CTR_Band'] = df['Ad_CTR'].apply(ctr_band)
```

```python
df[['Revenue', 'Ad_Spend', 'Effective_Revenue', 'Revenue_per_Unit', 'Discount_Per_Unit', 'Ad_CTR', 'CTR_Band']].head()
```

#### Weekend vs Weekday Transactions

```python
df['DayOfWeek'] = df['Transaction_Date'].dt.dayofweek
df['Is_Weekend'] = df['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)
```

```python
df[['Transaction_Date', 'DayOfWeek', 'Is_Weekend']].head()
```

```python
df.groupby('Is_Weekend')['Revenue'].sum()
```

####  Ad_Spend_Per_Click = Ad_Spend / Clicks

```python
df['Ad_Spend_Per_Click'] = df['Ad_Spend'] / df['Clicks'].replace(0, pd.NA)
```

```python
df[['Ad_Spend', 'Clicks', 'Ad_Spend_Per_Click']].head()
```

```python
df.sort_values(by='Ad_Spend_Per_Click', ascending=False).head(5)
```

#### Revenue to Discount Ratio

```python
df['Revenue_to_Discount_Ratio'] = df['Revenue'] / df['Discount_Applied'].replace(0, pd.NA)
```

```python
df[['Revenue', 'Discount_Applied', 'Revenue_to_Discount_Ratio']].head()
```

#### Conversion Rate

```python
# Conversion Rate = Clicks / Impressions
df['Conversion_Rate'] = df['Clicks'] / df['Impressions'].replace(0, pd.NA)
```

```python
df['Conversion_Rate']
```

#### Identify Category with highest Conversion_Rate

```python
category_conversion = df.groupby('Category')['Conversion_Rate'].mean().sort_values(ascending=False)
top_category = category_conversion.idxmax()

print("Category with highest Conversion Rate:", top_category)
```

#### Find customers (Customer_ID) with >3 transactions

```python
frequent_customers = df['Customer_ID'].value_counts()
customers_over_3 = frequent_customers[frequent_customers > 3].index.tolist()
print("Customers with >3 transactions:", customers_over_3)
```

#### Highlight any Region with high Ad_Spend but low Revenue

```python
# Calculate median values
median_ad_spend = df['Ad_Spend'].median()
median_revenue = df['Revenue'].median()

# Group by region
region_summary = df.groupby('Region')[['Ad_Spend', 'Revenue']].sum().reset_index()

# Filter regions with high ad spend but low revenue
target_regions = region_summary[
    (region_summary['Ad_Spend'] > median_ad_spend) & 
    (region_summary['Revenue'] < median_revenue)
]

print("Regions with high Ad_Spend but low Revenue:")
print(target_regions)
```

```python
region_summary['Revenue_per_AdSpend'] = region_summary['Revenue'] / region_summary['Ad_Spend'].replace(0, pd.NA)
region_summary.sort_values(by='Revenue_per_AdSpend')
```

 #### Top 3 Categories on Weekends vs Weekdays

# Visualization and Summary

#### Bar chart: Category vs Total Revenue

```python
import seaborn as sns

# Aggregate revenue by category
category_revenue = df.groupby('Category')['Revenue'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Revenue', data=category_revenue)
plt.title('Total Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Line chart: Transactions per day

```python
transactions_per_day = df['Date'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
transactions_per_day.plot(kind='line', marker='o', color='skyblue')
plt.title('Transactions Per Day')
plt.xlabel('Date')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Horizontal bar: Top 5 products by revenue

```python
top5_products = df.groupby('Product_ID')['Revenue'].sum().sort_values(ascending=True).tail(5)

plt.figure(figsize=(10, 6))
top5_products.plot(kind='barh', color='Red')
plt.title('Top 5 Products by Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Product ID')
plt.tight_layout()
plt.show()
```

#### Heatmap (optional): Correlation between Ad_CTR, Ad_CPC, Revenue,Clicks

```python
corr_features = df[['Ad_CTR', 'Ad_CPC', 'Revenue', 'Clicks']]
corr_matrix = corr_features.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

##### Revenue to Discount Ratio by Category

```python
disc_eff = df.groupby('Category')['Revenue_to_Discount_Ratio'].mean().sort_values(ascending=False).reset_index()

sns.barplot(x='Revenue_to_Discount_Ratio', y='Category', data=disc_eff, palette='viridis')
plt.title('Discount Efficiency by Category')
plt.xlabel('Revenue / Discount Applied')
plt.tight_layout()
plt.show()
```

#### Scatter Plot: Ad Spend vs Revenue

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Ad_Spend', y='Revenue', hue='Category', data=df, alpha=0.7)
plt.title('Ad Spend vs Revenue by Category')
plt.tight_layout()
plt.show()
```

#### Box Plot: Revenue per Unit by Category

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Revenue_per_Unit', data=df)
plt.xticks(rotation=45)
plt.title('Revenue Per Unit Distribution by Category')
plt.tight_layout()
plt.show()
```
